# Time Series Analysis of NAICS

## I. Preparation of datasets


### NAICS 2

```python
import pandas as pd
files_list1 = ['RTRA_Employ_2NAICS_97_99.csv','RTRA_Employ_2NAICS_00_05.csv','RTRA_Employ_2NAICS_06_10.csv','RTRA_Employ_2NAICS_11_15.csv','RTRA_Employ_2NAICS_16_20.csv']
naics2_df = pd.concat(map(pd.read_csv,files_list1), ignore_index=True)
naics2_df['naics_code'] = naics2_df.NAICS.str.split('[').map(lambda x: x[1].strip(']'))
naics2_df.shape
```

```python

naics2_1=naics2_df[naics2_df.naics_code.isin(['55','56'])].groupby(['SYEAR','SMTH'], as_index=False).agg('sum')
naics2_1['naics_code']='55 & 56'

naics2_final = pd.concat([naics2_df,naics2_1], ignore_index=True)

naics_code_list = ['22','23','41','53','81','55 & 56']

final_naics2_df= naics2_final[naics2_final.naics_code.isin(naics_code_list)]
final_naics2_df= final_naics2_df.drop(columns=["NAICS"])
```

### NAICS 3

```python
files_list2 = ['RTRA_Employ_3NAICS_97_99.csv','RTRA_Employ_3NAICS_00_05.csv','RTRA_Employ_3NAICS_06_10.csv','RTRA_Employ_3NAICS_11_15.csv','RTRA_Employ_3NAICS_16_20.csv']
naics3_df = pd.concat(map(pd.read_csv,files_list2), ignore_index=True)
naics3_df['naics_code'] = naics3_df.NAICS.map(lambda x: x.split('[')[-1].strip(']'))
naics3_df.naics_code.replace('Securities, commodity contracts, and other financial investment and related acti','523', inplace=True)
naics3_df.naics_code.replace('Other','81', inplace=True)
```

```python
naics_code_list2 = ['111 & 112','114','113','115','211','213','212','311 & 312','321', '322', '331',
                    '332', '333','313, 314, 315, 316, 323, 324, 325, 326, 327, 334, 335, 337 & 339','441', 
                    '446','442,443,444,445,446,447,448,451,452,453 & 454','481', '482', '483', '484', '488', 
                    '491 & 492', '485,486 & 487', '493', '524','521, 522, 523 & 526','621', '622', '623', '624',
                    '511', '512', '517','515, 518 & 519', '711', '713', '712', '721', '722', '911', '912',
                    '913, 914 & 919']


dictionnary = {
    '313, 314, 315, 316, 323, 324, 325, 326, 327, 334, 335, 337 & 339':['313', '314', '315', '316', '323', '324', '325', '326', '327', '334', '335', '337','339'],
    '111 & 112':['111','112'],
     '311 & 312':['311','312'],
    '442,443,444,445,446,447,448,451,452,453 & 454':['442','443','444','445','446','447','448','451','452','453','454'],
    '491 & 492':['491','492'],
    '485,486 & 487':['485','486','487'],
    '521, 522, 523 & 526':['521', '522','523','526'],
    '515, 518 & 519':['515', '518','519'],
    '913, 914 & 919':['913', '914', '919']
}
dataframe = [naics3_df,]

for key in dictionnary.keys():
    naics3_1=naics3_df[naics3_df.naics_code.isin(dictionnary[key])].groupby(['SYEAR','SMTH'], as_index=False).agg('sum')
    naics3_1['naics_code'] = key
    dataframe.append(naics3_1)

naics3_final = pd.concat(dataframe, ignore_index=True)

final_naics3_df= naics3_final[naics3_final.naics_code.isin(naics_code_list2)]
final_naics3_df.drop(columns="NAICS", inplace=True)
```

### NAICS 4


```python
files_list3 = ['RTRA_Employ_4NAICS_97_99.csv','RTRA_Employ_4NAICS_00_05.csv','RTRA_Employ_4NAICS_06_10.csv','RTRA_Employ_4NAICS_11_15.csv','RTRA_Employ_4NAICS_16_20.csv']
naics4_df = pd.concat(map(pd.read_csv,files_list3), ignore_index=True)
naics4_df.NAICS = naics4_df.NAICS.astype('str')
naics4_df.rename(columns={"NAICS":"naics_code"}, inplace=True)

list3 = ['3366','3361,  3362,  3363, 3364, 3365, 3369', '5413', '5415', '5416','5411, 5412, 5414, 5417, 5418 & 5419','6111', '6112', '6113','6114, 6115, 6116 & 6117']
dict1 = {
    '3361,  3362,  3363, 3364, 3365, 3369':['3361',  '3362',  '3363', '3364', '3365', '3369'],
    '5411, 5412, 5414, 5417, 5418 & 5419':['5411', '5412', '5414', '5417', '5418', '5419'],
    '6114, 6115, 6116 & 6117':['6114', '6115', '6116', '6117']
}

dataframe = [naics4_df,]

for key in dict1.keys():
    naics4_1=naics4_df[naics4_df.naics_code.isin(dict1[key])].groupby(['SYEAR','SMTH'], as_index=False).agg('sum')
    naics4_1['naics_code'] = key
    dataframe.append(naics4_1)

naics4_final = pd.concat(dataframe, ignore_index=True)

final_naics4_df= naics4_final[naics4_final.naics_code.isin(list3)]

final_naics4_df
```
